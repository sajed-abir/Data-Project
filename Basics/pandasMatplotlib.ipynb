 **Matplotlib**: Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0, 5, 0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
monthly_counts = df.job_posted_month.value_counts()
monthly_counts = monthly_counts.sort_index()
```

```python
plt.plot(monthly_counts.index, monthly_counts.values)
```

```python
job_counts = df.job_title_short.value_counts()
job_counts = job_counts.sort_values(ascending=True)
```

```python
plt.bar(job_counts.index, job_counts.values)
plt.title('Job Counts')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
country_counts = df.job_country.value_counts().head()
country_counts = country_counts.sort_values(ascending=True)
```

```python
plt.bar(country_counts.index, country_counts.values)
```

```python
avg_salary = df.groupby('job_title_short')['salary_year_avg'].median()
avg_salary = avg_salary.sort_values(ascending=True)
```

```python
plt.barh(avg_salary.index, avg_salary.values)
plt.title('Average Salary by Job Title')
plt.xlabel('salary $$$')
```
